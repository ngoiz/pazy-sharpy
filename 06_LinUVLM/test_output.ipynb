```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['figure.figsize'] = (8, 6)
%config InlineBackend.figure_format = 'retina'
```

```python
output_path = './output/postprocess/'
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(14, 10))
cmap=plt.get_cmap('Blues')
nrm = plt.Normalize(vmin=20, vmax=60)

for u_inf in [30, 35, 40, 45, 50, 55, 60]:
    for nout in range(3):
        for nin in range(3):
            qs = 0.5 * 1.225 * u_inf ** 2 * 0.1 * 0.55
            yfreq = np.loadtxt(output_path + '/sharpy_uinf{:04g}_aero/in{:02g}_out{:02g}.txt'.format(u_inf, nin, nout))
            w_to_k = 0.05/u_inf
            ax[nout, nin].plot(yfreq[:, 0], yfreq[:, 1] / qs, c=cmap(nrm(u_inf)))
#             ax[nout, nin].plot(yfreq[:, 0] / w_to_k, yfreq[:, 2], c=cmap(nrm(u_inf)), ls='--')


            
plt.tight_layout()
```
